```python
#Mistyped Value (Negative) - step 1 - get median value of Tumor size without the outlier
median_value = classification_data['Tumor_Size'][classification_data['Tumor_Size'] >= 0].median()
```

```python
#Mistyped Value (Negative) - step 2 - funtion to repplace negative numbers with the median value
def replace_negative(value):
    if value < 0:
        return median_value
    else:
        return value
```

```python
#Mistyped Value (Negative) - step 3 - replace negative numbers with the median value by calling the user defined funtion
classification_data['Tumor_Size'] = classification_data['Tumor_Size'].apply(replace_negative)
```

```python
#fix outliers - step 1 - get the median value of column Age
median_age = classification_data['Age'].median()
```

```python
#fix outliers - step 2 - funtion to replace extreme numbers with the median value

#define the funtion
def outlier(value):
#check weather the value is a negative number
    if value < 0:
#replace value with median
        return median_age
#check weather the value is a larger than 100 (normal human age is less than 100)
    elif value > 100:
#replace value with median
        return median_age
    else:
#keep the value with no change
        return value
```

```python
#fix outliers - step 3 - replace extreme numbers with the median value  by calling the user defined function
classification_data['Age'] = classification_data['Age'].apply(outlier)
```

```python
#replace missing values(NaN)
#
#Age is filled median
classification_data['Age'].fillna(classification_data['Age'].median(), inplace=True)
#sex is filled with mode (most frequant value)
classification_data['Sex'].fillna(classification_data['Sex'].mode()[0], inplace=True)
#tumor size is filled with median
classification_data['Tumor_Size'].fillna(classification_data['Tumor_Size'].median(), inplace=True)
#filled with 0 , assuming that no nodes were examined
classification_data['Regional_Node_Examined'].fillna(0, inplace=True)
```

```python
#turns sex column values to string type(Object), get rid of the white spaces, and 1st letter to capital of each word
classification_data['Sex'] = classification_data['Sex'].astype(str).str.strip().str.capitalize()
#turns Mortality Status column values to string type(Object), get rid of the white spaces, and 1st letter to capital of each word
classification_data['Mortality_Status'] = classification_data['Mortality_Status'].astype(str).str.strip().str.capitalize()
```

```python
#to check if the cleaning and transforming has worked - summery of missing (NaN numbers)
classification_data.isnull().sum()
```

```python
#to check if the cleaning and transforming has worked-to check unique numbers
classification_data['Mortality_Status'].unique()
```

```python
#Renaming categorical values (labels) in variables (manually)
classification_data['Sex'] = classification_data['Sex'].map({'1': 1, 'Female': 0})
classification_data['T_Stage'] = classification_data['T_Stage'].map({'T1': 1, 'T2': 2, 'T3': 3, 'T4': 4})
classification_data['N_Stage'] = classification_data['N_Stage'].map({'N1': 1, 'N2': 2, 'N3': 3})
classification_data['6th_Stage'] = classification_data['6th_Stage'].map({'IIA': 1, 'IIIA': 2, 'IIIC': 3, 'IIB': 4, 'IIIB': 5})
classification_data['Differentiated'] = classification_data['Differentiated'].map({'Poorly differentiated': 0, 'Moderately differentiated': 1, 'Well differentiated': 2, 'Undifferentiated': 3})
classification_data['A_Stage'] = classification_data['A_Stage'].map({'Regional': 1, 'Distant': 2})
classification_data['Estrogen_Status'] = classification_data['Estrogen_Status'].map({'Positive': 1, 'Negative': 0})
classification_data['Progesterone_Status'] = classification_data['Progesterone_Status'].map({'Positive': 1, 'Negative': 0})
classification_data['Mortality_Status'] = classification_data['Mortality_Status'].map({'Alive': 0, 'Dead': 1})
```
